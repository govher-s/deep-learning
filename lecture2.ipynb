```python
#lecture 2 MIT
import tensorflow as tf
from tensorflow import keras

input = keras.Input(shape=(29,))
hiddenlayer = keras.layers.Dense(16, activation="relu")(input)
output = keras.layers.Dense(1, activation="sigmoid")(hiddenlayer)
model = keras.Model(input, output)
```
